## Tratamiento de outliers

Calculo de IRQ y límites

```python
#Calculo de irq y límites
irq = estadisticos.loc["75%"] - estadisticos.loc["25%"]
limite_inferior = estadisticos.loc["25%"] - irq*1.75
limite_superior = estadisticos.loc["75%"] + irq*1.75

#Eliminación de outliers
for col in columnas:
    outlier_li = train[col]<=limite_inferior[col]
    outlier_ls = train[col]>=limite_superior[col]
    print(col, ", outliers tratados: ", sum(outlier_li), ", ", sum(outlier_ls))
    train.loc[outlier_li, col] = estadisticos.loc["mean",col]
    train.loc[outlier_ls, col] = estadisticos.loc["50%",col]
```

## Feature Scaling

Normalización de variables

```python
train = (train - estadisticos.loc["mean"])/estadisticos.loc["std"]
```

# Tensorflow

Función auxiliar

```python
def my_reshape(my_array):
    if( len(my_array.shape)==1):
        return my_array.reshape(my_array.shape[0], 1)
    else:
        return my_array
```

```python
class ModeloLineal:
            
    def __init__(self, x, y, tamaño_muestra, lr, epochs, batch_size, mb, string, iteraciones):
        tf.reset_default_graph()
        self.g = tf.Graph()
        
        #Definición de variables a usar en la clase
        x_temporal = my_reshape(x)
        self.x = np.hstack((x_temporal , np.ones((x_temporal.shape[0],1)) ))
        self.lr = lr
        self.y = y
        self.mb = mb
        self.mb_len = len(mb)
        self.epochs = epochs
        self.tamaño_muestra = tamaño_muestra
        self.batch_size = batch_size
        self.string = string
        self.iteraciones = iteraciones
        
    def predecir(self):
        
        #Grafo
        with self.g.as_default():
            
            #Definición de variables
            mb = tf.Variable(np.zeros((self.mb_len, 1)), dtype=tf.dtypes.float64, name="mb")
            lr = tf.Variable(0.0, name="lr", dtype=tf.dtypes.float64)
            x = tf.Variable(tf.zeros((self.batch_size, self.mb_len), dtype=tf.dtypes.float64), name="x")
            y = tf.Variable(tf.zeros((self.batch_size, 1), dtype=tf.dtypes.float64), name="y")
            
            #hipotesis
            with tf.name_scope("modelo"):
                y_estimado = tf.matmul(x, mb, name="y_estimado")
            
            #Cálculo de error
            with tf.name_scope("error"):
                error = tf.math.multiply(
                    tf.cast(1/2, dtype=tf.dtypes.float64),
                    tf.reduce_mean(
                        tf.math.square(
                            tf.math.subtract(
                                y, 
                                y_estimado
                            )
                        ) 
                    )
                )
            
            #Registro de error
            error_summary = tf.summary.scalar("error",error)
            
            #Gradiente
            with tf.name_scope("gradiente"):
                grad_mb = tf.gradients(error,[mb])
            
            #Siguiente paso
            with tf.name_scope("siguiente"):
                mb_final = tf.subtract(mb, tf.multiply(lr,grad_mb[0]))
        
        #Sesión
        with tf.Session(graph=self.g) as sesion:
            #Escritura de log
            writer = tf.summary.FileWriter(
                os.path.join(
                    "./logs-parte2", 
                    self.string+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
                ), 
                sesion.graph
            )
            
            #Epochs
            for epoch in range(self.epochs):
                #Iteraciones
                for iteracion in range(self.iteraciones):
                    #Tamaño de muestra
                    muestra_inicio =int(tamaño_muestra/epochs)*epoch
                    muestra_fin = muestra_inicio+self.batch_size

                    #TensorFlow
                    [err,self.mb,es] = sesion.run([error, mb_final,error_summary], feed_dict = {
                        mb: self.mb,
                        x:np.array(self.x[muestra_inicio:muestra_fin,:]).reshape([-1,self.mb_len]),
                        y:np.array(self.y[muestra_inicio:muestra_fin]).reshape([-1,1]),
                        lr: self.lr
                    }) 

                    #Escritura de resultados
                    writer.add_summary(es, epoch*epochs + iteracion)
                
            writer.close()
            
        return [self.mb]
```

# Análisis de resultados

Las siguientes funciones realizan la predicción de datos con TF y con Scikit Learn.

```python
def scikitlearn_regresion_lineal(x, y):
    scikitlearn_modelo = LinearRegression().fit(x, y)
    return scikitlearn_modelo

def predecir_modelo_tf(x, mb):
    i=0
    for col in columnas:
        x[i] = (x[i] - estadisticos.loc["mean", col])/estadisticos.loc["std", col]
        i+=1
    matriz_auxiliar = np.hstack((x, np.ones((x.shape[0], 1)) ))
    r = np.matmul(matriz_auxiliar, mb)
    r = r*estadisticos.loc["std","quality"] + estadisticos.loc["mean","quality"]
    return r

def calcular_estimacion(scikitlearn, mb_tf1, mb_tf2, mb_tf3, mb_tf4):
    scikitlearn_prediccion = scikitlearn[0].predict(scikitlearn[1])
    modelo_tf1 = predecir_modelo_tf(mb_tf1[1], mb_tf1[0])
    modelo_tf2 = predecir_modelo_tf(mb_tf2[1], mb_tf2[0])
    modelo_tf3 = predecir_modelo_tf(mb_tf3[1], mb_tf3[0])
    modelo_tf4 = predecir_modelo_tf(mb_tf4[1], mb_tf4[0])
    promedio = np.nanmean(np.array([scikitlearn_prediccion, modelo_tf1, modelo_tf2, modelo_tf3, modelo_tf4]), axis=0)
    return (scikitlearn_prediccion, modelo_tf1, modelo_tf2, modelo_tf3, modelo_tf4, promedio)
```

Calculando:

```python
x_real = my_reshape(data_validation[columnas].to_numpy())
y_real =  my_reshape(data_validation[columna_y].to_numpy())

x_training_sl = my_reshape(data_training[columnas].to_numpy())
y_training_sl = my_reshape(data_training[columna_y].to_numpy())

scikitlearn = scikitlearn_regresion_lineal(x_training_sl, y_training_sl)
estimacion = calcular_estimacion(
    [scikitlearn, x_real], 
    [resultado1[0], x_real], 
    [resultado2[0], PolynomialFeatures(2).fit_transform(x_real)], 
    [resultado3[0], PolynomialFeatures(3).fit_transform(x_real)], 
    [resultado4[0], PolynomialFeatures(5).fit_transform(x_real)]) 
```

Comparación con gráfica de barras

```python
def grafica_barras(scikitlearn, modelo_tf1, modelo_tf2,modelo_tf3,modelo_tf4, promedio, titulo):
    plt.rcParams["figure.figsize"] = (4, 3)
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    titulos = ['scikitlearn', 'grado1', 'grado2', 'grado3', 'grado4', 'promedio']
    error = [
        np.nanmean(scikitlearn), 
        np.nanmean(modelo_tf1), 
        np.nanmean(modelo_tf2), 
        np.nanmean(modelo_tf3), 
        np.nanmean(modelo_tf4), 
        np.nanmean(promedio)
    ]
    print(error)
    ax.bar(titulos,error)
    plt.title(titulo)
    plt.show()

#Errores para la primera variable con el valor más predictivo
error_scikitlearn = y_real - estimacion[0]
error_modelo_tf1 = y_real - estimacion[1]
error_modelo_tf2 = y_real - estimacion[2]
error_modelo_tf3 = y_real - estimacion[3]
error_modelo_tf4 = y_real - estimacion[4]
error_promedio = y_real - estimacion[5]
grafica_barras(error_scikitlearn, error_modelo_tf1, error_modelo_tf2, error_modelo_tf3, error_modelo_tf4, error_promedio, "Error")
```

Comparación con gráfica de líneas

```python
def grafica_lineas(precio, scikitlearn, modelo_tf, titulo, linea):
    index = np.arange(0, precio.shape[0], 1)
    
    plt.rcParams["figure.figsize"] = (12, 9)
    f, ax = plt.subplots(1)
    plt.title(titulo)
    plt.xlabel("Calidad")
    plt.ylabel('Muestra')
    ax.plot(index, precio)
    ax.plot(index, scikitlearn, color="green")
    ax.plot(index, modelo_tf, color="red")
    ax.legend(['Estimados real', 'Scikitlearn', linea])
    ax.set_ylim(ymin=0)
    plt.show(f)
    
grafica_lineas(y_real, estimacion[0], estimacion[1], "Comparación de resultados", "Grado1")
```
